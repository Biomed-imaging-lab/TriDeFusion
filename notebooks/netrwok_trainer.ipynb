# Training journal of the denoising network 
The journal is necessary for network training. The training parameters are specified for input in the input data block:

- *DATASET_PATH* - path to the directory with the dataset
- *EPOCHS* - number of epochs
- *LR* - learning rate
- *BATCH_SIZE* - batch size
- *MODEL_NAME* - model name
- *TRAINING_TYPE* - training type (classic or distillation)

```python
DATASET_PATH = "./dataset"
EPOCHS = 100
LR = 1e-4
BATCH_SIZE = 4
MODEL_NAME = "fluoro-msa"
```

```python
!run -i train.py --exp-name=$MODEL_NAME --data-root=$DATASET_PATH --lr=$LR --epochs=$EPOCHS --batch-size=$BATCH_SIZE
```
